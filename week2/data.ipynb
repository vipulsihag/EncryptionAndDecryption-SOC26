```python
# Reading Data
with open("originalData.txt", "r") as file:
    text = file.read()
chars = {}
text = text.lower()
for ch in text:
    if not ch in chars.keys(): chars[ch] = 1
    else : chars[ch] += 1
# List of allowed characters
characters = []
characters = [chr(ord('a') + i) for i in range(26)]

# Take the 11 most common non alphabetic characters
for c, n in sorted(chars.items(), key= lambda x : x[1], reverse=True):
    if (len(characters) < 37):
        if c not in characters: characters.append(c)
    else : break

print(len(characters))
print(characters)
```

```python
# Processing the data, removing extra characters
processedText = ''.join(c for c in (text) if c in characters)
print(len(processedText))
```

### Encrypting the data using Caesar Cipher of various shifts

```python
# Some Parameters
numChars = len(characters)
segmentSize = 500 # 400 for training, 100 for testing
numOfSegments = int(len(processedText)/segmentSize)
inputWindow = 50 # Model given input as 50 characters one at a time
```

```python
def apply_shift(text, shift, characters):
    l = len(characters)
    lst = list(text)
    for i in range(len(lst)):
        char = lst[i]
        j = 0
        while(characters[j] != char):
            j += 1
        lst[i] = characters[(j+shift)%l]
    return ''.join(lst)
```

```python
import random
label = open("encryptionLabels.csv", "w")
label.write("train,test\n")

train = open("trainData.txt", "w")
test = open("testData.txt", "w")

for i in range(numOfSegments):
    shiftTrain = random.randint(0,numChars-1)
    shiftTest = random.randint(0,numChars-1)
    label.write(str(shiftTrain)+","+str(shiftTest)+"\n")
    train.write(apply_shift(processedText[segmentSize*i : segmentSize*i + int(segmentSize*0.8)], shiftTrain, characters)+"|") # '|' is seperator
    test.write(apply_shift(processedText[segmentSize*i + int(segmentSize*0.8) : segmentSize*(i+1)], shiftTest, characters)+"|")

label.close()
train.close()
test.close()
```
